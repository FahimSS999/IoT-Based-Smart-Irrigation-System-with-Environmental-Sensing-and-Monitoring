IQR


```python
numerical_cols = x.columns

def detect_outliers_iqr(data, column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)

  IQR = Q3 - Q1
  Lower = Q1 - 1.5*IQR
  Upper = Q3 + 1.5*IQR

  outliers = data[(data[column]<Lower) | (data[column] > Upper)]
  return outliers

for col in numerical_cols:
  outliers = detect_outliers_iqr(x, col)
  print(f"\n Outliers in '{col}' : {len(outliers)} rows")
  print(outliers.head())
```

Z Score

```python
from scipy.stats import zscore
import numpy as np
import pandas as pd

z_scores = zscore(x)
z_scores_df = pd.DataFrame(z_scores, columns=x.columns)

formatted_df = x.copy()

for col in x.columns:
    formatted_df[col] = x[col].astype(str) + " (" + z_scores_df[col].round(2).astype(str) + ")"


print("\nx with Z-scores in brackets:\n")
print(formatted_df.head())
```
